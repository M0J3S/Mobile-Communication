```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_subcar = 64
num_symbol = 5
cp_len = 16
mod_order = 4
```

```python
data = np.random.randint(0,mod_order, size= (num_subcar, num_symbol))
qam_symbol = np.exp(1j*2*np.pi*data/mod_order)
```

```python
ifft_output = np.fft.ifft(qam_symbol, axis=1)
```

```python
cyclic_prifix = ifft_output[:,-cp_len:]
ofdm_symbol_with_symbol_len = np.hstack([cyclic_prifix,ifft_output])
```

```python
ofdm_sig = ofdm_symbol_with_symbol_len.flatten()
```

```python
plt.plot(np.real(ofdm_sig))
plt.plot(np.imag(ofdm_sig))
```
