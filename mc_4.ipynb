```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = 72
vm = (v*1000)/3600
vm
```

```python
c=3e8
f= 900e6
```

```python
lambda_c = c/f
lambda_c
```

```python
theta = np.array([180,0,60,90])
theta_rad = np.deg2rad(theta)
fd = (vm*np.cos(theta_rad)/lambda_c)
```

```python
fr = f + fd
```

```python
plt.plot(theta,fd)
```

```python
plt.plot(fr,fd)
```
